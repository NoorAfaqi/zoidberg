```python
import os
import shutil

# Base dataset path
base_path = "datasets/chest_Xray"

# Splits
splits = ["train", "test", "val"]

for split in splits:
    pneumonia_path = os.path.join(base_path, split, "PNEUMONIA")

    virus_path = os.path.join(base_path, split, "VIRAL_PNEUMONIA")
    bacteria_path = os.path.join(base_path, split, "BACTERIAL_PNEUMONIA")

    os.makedirs(virus_path, exist_ok=True)
    os.makedirs(bacteria_path, exist_ok=True)

    for filename in os.listdir(pneumonia_path):
        file_path = os.path.join(pneumonia_path, filename)

        if not os.path.isfile(file_path):
            continue

        if "virus" in filename.lower():
            dest_path = os.path.join(virus_path, filename)
        elif "bacteria" in filename.lower():
            dest_path = os.path.join(bacteria_path, filename)
        else:
            print(f"Skipping unknown file: {filename}")
            continue

        # Move file
        shutil.move(file_path, dest_path)

    print(f"{split} split processed.")

print("Done!")
```
